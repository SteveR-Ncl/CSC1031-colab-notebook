# Example functions and their graphs


## $y = x^2$
An example graph of $y=x^2$.

```python
x=np.linspace(-10,10,100)
plt.plot(x, x**2)
plt.grid()
plt.title("$y=x^2$")
```

## $y=x+1$

```python
x=np.linspace(-4,4)
plt.plot(x, x+1)
plt.grid()
plt.title("$y=x+1$")
```

## $y = \sqrt{x}+1$


```python
x = np.linspace(0,50)
y = np.sqrt(x)+1
plt.plot(x,y)
plt.grid()
plt.title("$y = \sqrt{x}+1$")
```

## Domain and range
Square root inverse, $y=1/\sqrt(x+1)$

Domain for this function is $(-1,\infty ] $

To avoid overflow in drawing the graph, we specify `endpoint=False` and write the interval as `5, -1` so that the endpoint is applied to `-1`.

```python
x = np.linspace(5,-1,endpoint=False)
y = 1/np.sqrt(x+1)
plt.plot(x,y)
plt.grid()
plt.title("$y = 1/\sqrt{x+1}$")
```

```python
x=np.linspace(-2,3,6)
y=1+x**2
x1=np.linspace(-2,3)
y1=1+x1**2
plt.plot(x1,y1)
plt.plot(x,y,'o')
plt.grid()
plt.title("Sketched Graph $y=1+x^2$")
```


## Absolute value $y = |x|$

  $$
  |x| = \left\{
  \begin{array}\\
  x & \mbox{if } 0 \le x < \infty \\
  -x & \mbox{if} -\infty < x < 0
  \end{array}
  \right. 
  $$

 Graph of $y=|x|$

```python
x=np.linspace(-50,50,500)
y=abs(x)
plt.plot(x,y)
plt.title("Modulus")
```

# Quadratic equations


## Solving a quadtratic 

Height of a ball is given by $h(t) = 15t-5t^2$

At what time t is the height = 10?


```python
t=np.linspace(0,3)
h=15*t - 5*t**2
plt.plot(t,h)
plt.grid()
t1 = np.linspace(1,2,2)
h1 = 15*t1 - 5*t1**2
plt.plot(t1,h1,"o")
plt.title("$h(t) = 15t - 5t^2$")

```
